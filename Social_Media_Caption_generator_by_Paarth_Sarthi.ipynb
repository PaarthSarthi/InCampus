```python
# STEP 1: Install dependencies
!pip install transformers torch --quiet

# STEP 2: Import libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# STEP 3: Load GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium" or others for better results
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()

# STEP 4: Define caption generator function
def generate_caption(prompt, max_length=50, temperature=0.8, top_k=50):
    inputs = tokenizer.encode(prompt, return_tensors='pt')
    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_length=max_length,
            temperature=temperature,
            top_k=top_k,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )
    generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated

# STEP 5: Try it!
prompt = "A sunny beach with palm trees - Instagram caption:"
caption = generate_caption(prompt)
print("\nGenerated Caption:\n", caption)
```

```python
# STEP 1: Install required packages
!pip install transformers gradio git+https://github.com/salesforce/BLIP.git -q
!pip install torchvision --quiet

# STEP 2: Import necessary libraries
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import gradio as gr
import random

# STEP 3: Load GPT-2 model for caption generation
caption_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
caption_model = GPT2LMHeadModel.from_pretrained("gpt2")
caption_model.eval()

# STEP 4: Load BLIP for image-to-text (image captioning)
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# STEP 5: Caption generator function
def generate_caption(prompt, max_length=50, temperature=0.8, top_k=50):
    inputs = caption_tokenizer.encode(prompt, return_tensors='pt')
    with torch.no_grad():
        outputs = caption_model.generate(
            inputs,
            max_length=max_length,
            temperature=temperature,
            top_k=top_k,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            do_sample=True,
            pad_token_id=caption_tokenizer.eos_token_id
        )
    return caption_tokenizer.decode(outputs[0], skip_special_tokens=True)

# STEP 6: Convert image to descriptive prompt using BLIP
def image_to_text(image):
    inputs = blip_processor(image, return_tensors="pt")
    out = blip_model.generate(**inputs)
    description = blip_processor.decode(out[0], skip_special_tokens=True)
    return description

# STEP 7: Hashtag generator (simple random mockup)
def generate_hashtags(description, count=5):
    base_tags = [
        "#vibes", "#travel", "#nature", "#fun", "#love", "#happy", "#sunset", "#photography",
        "#style", "#lifestyle", "#wanderlust", "#instamood", "#art", "#smile"
    ]
    random.seed(hash(description) % 10000)  # Seeded for consistent output
    return ' '.join(random.sample(base_tags, count))

# STEP 8: Unified Gradio Interface
def generate_full_output(image, text_prompt, style):
    if image is not None:
        prompt = image_to_text(image)
    elif text_prompt:
        prompt = text_prompt
    else:
        return "Please upload an image or enter a text prompt.", ""

    styled_prompt = f"{prompt.strip()} - Write a {style} social media caption:"
    caption = generate_caption(styled_prompt)
    hashtags = generate_hashtags(prompt)

    return caption.strip(), hashtags

# STEP 9: Launch Gradio app
gr.Interface(
    fn=generate_full_output,
    inputs=[
        gr.Image(type="pil", label="Upload Image (optional)"),
        gr.Textbox(label="Or enter a text prompt"),
        gr.Radio(["funny", "professional", "poetic", "motivational", "witty"], label="Choose Tone")
    ],
    outputs=[
        gr.Textbox(label="Generated Caption"),
        gr.Textbox(label="Suggested Hashtags")
    ],
    title="AI Social Media Caption Generator",
    description="Upload an image or type a prompt. Choose a style and get a caption + hashtags!",
    allow_flagging="never"
).launch()
```
